# Solution Code



```python
%matplotlib inline
import numpy as np

xs = [-1, 1, 2, 3, 4]
ys = [0, 3, 4, 7, 10]

def beta(xs, ys):
    numerator = 0.
    denominator = 0.
    x_mean = np.mean(xs)
    y_mean = np.mean(ys)
    for x, y in zip(xs, ys):
        numerator += (x - x_mean) * (y - y_mean)
        denominator += (x - x_mean) * (x - x_mean)
    return numerator / denominator

def alpha(xs, ys):
    b = beta(xs, ys)
    a = np.mean(ys) - b * np.mean(xs)
    return a

# Plotting

from matplotlib import pyplot as plt

def predict(xs, ys):
    a = alpha(xs, ys)
    b = beta(xs, ys)
    yhats = [a + b * x for x in xs]
    return yhats

ps = predict(xs, ys)

# The raw data
plt.scatter(xs, ys)
# The best fit line
plt.plot(xs, ps)
plt.show()
```
