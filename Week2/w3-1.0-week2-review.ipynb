## Matplotlib Sub-plotting Reference

I wanted to share some multi-plot figures so you can start thinking about doing these things in your own work to save time / space when rendering plots for presentation.

Generally, it's easier to control the layout and visual aspects **from a granular** perspective, using matplotlib. 

**Regarding the ".subplot(N)" notation:**
> In the case when *nrows*, *ncols* and *plot_number* are all less than 10,
> a convenience exists, such that the a 3 digit number can be given instead,
> where the hundreds represent *nrows*, the tens represent *ncols* and the
> units represent *plot_number*. 

>  For instance:

>>  subplot(211) # 211, 2 = nrows, 1 = ncols, 1 = plot number (when dealing with < 10 items)

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

# Grouping "The Hard Way"

We create 3 seperate DataFrames, with the same column "A", but having different categorical variable "animal_category", to show the category of animal.

````python
n_records = 10 # will control the size of our series

df_goats = pd.DataFrame({"a": [1] * n_records, "animal_category": "goats"})
df_cats  = pd.DataFrame({"a": [2] * n_records, "animal_category": "cats"})
df_ducks = pd.DataFrame({"a": [3] * n_records, "animal_category": "ducks"})
````

````python
from IPython.display import display, HTML
HTML("""
<table>
    <!-- I know this isn't kosher HTML, but it works though! -->
    <tr>
        <td><h3>df['animal_category'] == 'goats'</h3></td>
        <td>%s</td>
    </tr>
    <tr>
        <td><h3>df['animal_category'] == 'cats'</td>
        <td>%s</td>
    </tr>
    <tr>
        <td><h3>df['animal_category'] == 'ducks'</td>
        <td>%s</td>
    </tr>
</table>
""" % (df_goats.to_html(), df_cats.to_html(), df_ducks.to_html()))
````

# What if we just wanted to combine these into one dataset?

````python
# We use concat
animal_df = pd.concat([df_goats, df_cats, df_ducks]) ## Add goats, cats, and duck related dataframes into multi-animal dataframe
animal_df
````

## There is an easier way to do this! It's called `.grouby()`

````python
animal_df.groupby("animal_category").mean()
````

## Last thing about groups:

 - grouby() only segments your groups
 - a `groupby datafarme` object does not print a table by itself (but you can use the .groups property to see which groups exist)
 - You must aggregate something -- IE: a method that operates on multiple values.
 - You can specify a single column, but if you don't, the groupby will hand off all columns to the aggregation function you send it.
 - The most common function you will perform with groups is "aggregation"
 - `.size()` vs `.count()` is much and will give you different values depending on your data.
   - When you want to know number of records in a group, use `.size()`
   - When you want to know something about non-null, or null values in all columns in a group, use `.count()`
 - You can group by more than one variable and you will want to when you have many categorical values
   - You wouldn't know which subgroups may differ from global summary statistics or the latent characterists behind what is known "overall".
     - Which groups have the best performance?
     - Which regions are making the most money?
     - Which roles in a company have the highest expenses?
     - Which doors in a storage facility are the widest **VS** how many boxes must fit **VS** how many boxes _can_ fit ?
