```python
# import modules

import pandas as pd
import numpy as np
```

```python
# create a dataframe

data = {'name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'age': [42, 52, 36, 24, 73], 
        'preTestScore': [4, 24, 31, 2, 3],
        'postTestScore': [25, 94, 57, 62, 70]}
df = pd.DataFrame(data, columns = 
                  ['name', 'age', 'preTestScore', 'postTestScore'])
df
```

```python
# select the preTestScore column
df['preTestScore']
```

```python
# find the standard deviation of preTestScore 

df.preTestScore.std()
```

```python
# select rows in preTestScore with scores greater than the mean
# find the mean of preTestScore
pTS_mean = df['preTestScore'].mean() # Find mean of the preTestScore Column
pTS_mean
```

```python
# select the rows in preTestScore with scores greater than the mean

df[df['preTestScore']> pTS_mean] # select all columns in DF that are greater than the mean
```

```python
# create a new column that shows the score change from 
# preTestScore to postTestScore

df['TestDiff'] = (df['postTestScore']-df['preTestScore'])
```
